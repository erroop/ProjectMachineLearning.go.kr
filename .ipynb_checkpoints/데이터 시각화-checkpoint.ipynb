```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sms
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
```

```python
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
gt = pd.read_csv('./교통사고_날씨.csv', encoding = 'cp949', index_col = 0)
```

```python
gt.head(1)
```

```python
gt['사상자수'] = gt['사망자수'] + gt['중상자수']
```

```python
### 도로형태

#### 단일로에서 사고가 제일 많이 발생한 것을 확인 할 수 있다

plt.subplot(2, 2, 1)
plt.subplots_adjust(hspace=1)
plt.title('도로형태별 사고 발생 수')
gt['도로형태'].value_counts().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 2)
plt.title('도로형태별 사망자 수')
gt.groupby('도로형태')['사망자수'].count().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 3)
plt.title('도로형태별 중상자 수')
gt.groupby('도로형태')['중상자수'].count().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 4) 
plt.title('도로형태별 사상자 수')
gt.groupby('도로형태')['사상자수'].count().plot(kind = 'bar', figsize=(15,15))
```

```python
### 사고유형

#### 기타요인과 횡단중 사고가 많은 것을 확인 할 수 있다

plt.subplot(2, 2, 1)
plt.subplots_adjust(hspace=1)
plt.title('사고유형별 사고 발생 수')
gt['사고유형_세부'].value_counts().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 2)
plt.title('사고유형별 사망자 수')
gt.groupby('사고유형_세부')['사망자수'].count().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 3)
plt.title('사고유형별 중상자 수')
gt.groupby('사고유형_세부')['중상자수'].count().plot(kind = 'bar', figsize=(15,15) )

plt.subplot(2, 2, 4) 
plt.title('사고유형별 사상자 수')
gt.groupby('사고유형_세부')['사상자수'].count().plot(kind = 'bar', figsize=(15,15))
```

```python
### 가해자 법규 위반

#### 안전운전 의무 불이행으로 인한 사고가 제일 많은 것을 볼 수 있다

plt.subplot(2, 2, 1)
plt.subplots_adjust(hspace=1)
plt.title('가해자 법규 위반별 사고 발생 수')
gt['가해자법규위반'].value_counts().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 2)
plt.title('가해자 법규 위반별 사망자 수')
gt.groupby('가해자법규위반')['사망자수'].count().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 3)
plt.title('가해자 법규 위반별 중상자 수')
gt.groupby('가해자법규위반')['중상자수'].count().plot(kind = 'bar', figsize=(15,15) )

plt.subplot(2, 2, 4) 
plt.title('가해자 법규 위반별 사상자 수')
gt.groupby('가해자법규위반')['사상자수'].count().plot(kind = 'bar', figsize=(15,15))
```

```python
### 주/야   

#### 별차이 없으나 야간에 사고가 좀 더 많음

plt.subplot(2, 2, 1)
plt.subplots_adjust(hspace=1)
plt.title('주 / 야간별 사고 발생 수')
gt['주야'].value_counts().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 2)
plt.title('주 / 야간별 사망자 수')
gt.groupby('주야')['사망자수'].count().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 3)
plt.title('주 / 야간별 중상자 수')
gt.groupby('주야')['중상자수'].count().plot(kind = 'bar', figsize=(15,15) )

plt.subplot(2, 2, 4) 
plt.title('주 / 야간별 사상자 수')
gt.groupby('주야')['사상자수'].count().plot(kind = 'bar', figsize=(15,15))
```

```python
### 요일

#### 금요일에 사고가 제일 많은 것을 확인 할 수 있다

plt.subplot(2, 2, 1)
plt.subplots_adjust(hspace=1)
plt.title('요일별 사고 발생 수')
gt['요일'].value_counts().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 2)
plt.title('요일별 사망자 수')
gt.groupby('요일')['사망자수'].count().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 3)
plt.title('요일별 중상자 수')
gt.groupby('요일')['중상자수'].count().plot(kind = 'bar', figsize=(15,15) )

plt.subplot(2, 2, 4) 
plt.title('요일별 사상자 수')
gt.groupby('요일')['사상자수'].count().plot(kind = 'bar', figsize=(15,15))
```

```python
### 시군별

#### 서울시가 사고가 제일 많은 것을 확인 할 수 있다

plt.subplot(4, 1, 1)
plt.subplots_adjust(hspace=1)
plt.title('시/ 군별 사고 발생 수')
gt['시군'].value_counts().plot(kind = 'bar', figsize=(20,50))

plt.subplot(4, 1, 2)
plt.title('시/ 군별 사망자 수')
gt.groupby('시군')['사망자수'].count().plot(kind = 'bar', figsize=(20,50))

plt.subplot(4, 1, 3)
plt.title('시/ 군별 중상자 수')
gt.groupby('시군')['중상자수'].count().plot(kind = 'bar', figsize=(20,50) )

plt.subplot(4, 1, 4) 
plt.title('시/ 군별 사상자 수')
gt.groupby('시군')['사상자수'].count().plot(kind = 'bar', figsize=(20,50))
```

```python
### 월별 

#### 10월달에 사고가 제일 많은 것을 확인 할 수 있다

plt.subplot(2, 2, 1)
plt.subplots_adjust(hspace=1)
plt.title('월별 사고 발생 수')
gt['월'].value_counts().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 2)
plt.title('월별 사망자 수')
gt.groupby('월')['사망자수'].count().plot(kind = 'bar', figsize=(15,15))

plt.subplot(2, 2, 3)
plt.title('월별 중상자 수')
gt.groupby('월')['중상자수'].count().plot(kind = 'bar', figsize=(15,15) )

plt.subplot(2, 2, 4) 
plt.title('월별 사상자 수')
gt.groupby('월')['사상자수'].count().plot(kind = 'bar', figsize=(15,15))
```

```python
### 사고가 제일 많은 달 도로형태별 월평균 기온 강수량 습도 풍속

gt_10 = gt[gt['월'] == 10]

plt.subplot(2, 2, 1)
plt.subplots_adjust(hspace=1)
plt.title('10월 사고발생한 도로형태별 평균기온')
gt_10.groupby('도로형태')['평균기온(°C)'].mean().plot(kind= 'bar' , figsize=(15,15))

plt.subplot(2, 2, 2)
plt.title('10월 사고발생한 도로형태별 평균강수량')
gt_10.groupby('도로형태')['일강수량(mm)'].mean().plot(kind= 'bar', figsize=(15,15))

plt.subplot(2, 2, 3)
plt.title('10월 사고발생한 도로형태별 평균풍속')
gt_10.groupby('도로형태')['평균 풍속(m/s)'].mean().plot(kind= 'bar', figsize=(15,15))

plt.subplot(2, 2, 4)
plt.title('10월 사고발생한 도로형태별 평균습도')
gt_10.groupby('도로형태')['평균 상대습도(%)'].mean().plot(kind= 'bar', figsize=(15,15))

```

```python
### 사고가 제일 적은 달 도로형태별 월평균 기온 강수량 습도 풍속

gt_2 = gt[gt['월'] == 2]

plt.subplot(2, 2, 1)
plt.subplots_adjust(hspace=1)
plt.title('2월 사고발생한 도로형태별 평균기온')
gt_2.groupby('도로형태')['평균기온(°C)'].mean().plot(kind= 'bar' , figsize=(15,15))

plt.subplot(2, 2, 2)
plt.title('2월 사고발생한 도로형태별 평균강수량')
gt_2.groupby('도로형태')['일강수량(mm)'].mean().plot(kind= 'bar', figsize=(15,15))

plt.subplot(2, 2, 3)
plt.title('2월 사고발생한 도로형태별 평균풍속')
gt_2.groupby('도로형태')['평균 풍속(m/s)'].mean().plot(kind= 'bar', figsize=(15,15))

plt.subplot(2, 2, 4)
plt.title('2월 사고발생한 도로형태별 평균습도')
gt_2.groupby('도로형태')['평균 상대습도(%)'].mean().plot(kind= 'bar', figsize=(15,15))
```

```python
### 사고가 제일 많은 달 사고유형별 월평균 기온 강수량 습도 풍속

gt_10 = gt[gt['월'] == 10]

plt.subplot(2, 2, 1)
plt.subplots_adjust(hspace=1)
plt.title('10월 사고발생한 도로형태별 평균기온')
gt_10.groupby('사고유형_세부')['평균기온(°C)'].mean().plot(kind= 'bar' , figsize=(15,15))

plt.subplot(2, 2, 2)
plt.title('10월 사고발생한 도로형태별 평균강수량')
gt_10.groupby('사고유형_세부')['일강수량(mm)'].mean().plot(kind= 'bar', figsize=(15,15))

plt.subplot(2, 2, 3)
plt.title('10월 사고발생한 도로형태별 평균풍속')
gt_10.groupby('사고유형_세부')['평균 풍속(m/s)'].mean().plot(kind= 'bar', figsize=(15,15))

plt.subplot(2, 2, 4)
plt.title('10월 사고발생한 도로형태별 평균습도')
gt_10.groupby('사고유형_세부')['평균 상대습도(%)'].mean().plot(kind= 'bar', figsize=(15,15))
```
