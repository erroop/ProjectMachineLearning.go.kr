```python
y=df['사망_중상자비율']
x=df.drop(['사망_중상자비율','사망자수','중상자수','총인수','부상자수',\
           '최저기온(°C)', '최고기온(°C)','최대 풍속(m/s)', '최소 상대습도(%)',\
          '시군', '시군_날씨'],axis=1)
```

```python
x=pd.get_dummies(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
lr = LinearRegression()  # 알고리즘 객체 생성
lr.fit(x_train, y_train)  # 알고리즘 train data로 학습 => (black box) => 모델
preds = lr.predict(x_test)  # 모델(lr) : predict

mse = mean_squared_error(y_test, preds)
rmse = np.sqrt(mse)
r2_s = r2_score(y_test, preds)

print('MSE:',mse)
print('RMSE:',mse, rmse, r2_s)
print('R2_score:',mse, rmse, r2_s)
```
