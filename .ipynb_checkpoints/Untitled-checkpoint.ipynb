```python
df=pd.read_csv('2012-2021_교통사망사고정보.csv',encoding='CP949', index_col=0)
```

```python
df['발생지시_군']=df['발생지시도']+' '+df['발생지시군구']
```

```python
L=[]
for d in df['발생지시_군']:
    a=d.split(' ')[0]
    b=d.split(' ')[1]
    if '시' in a:
        L.append(a)
    elif '광주' == a:
        L.append(a)
    elif '대구' == a:
        L.append(a)
    elif '대전' == a:
        L.append(a)
    elif '부산' == a:
        L.append(a)
    elif '서울' == a:
        L.append(a)
    elif '세종' == a:
        L.append(a)
    elif '울산' == a:
        L.append(a)
    elif '인천' == a:
        L.append(a)
    else:
        L.append(b)
```

```python
L=pd.DataFrame(L)
M=L[0].unique()
```

```python
for m in M:
    print(m)
```

```python
df['발생지시_군']=L
```

```python
df.drop(['발생지시도', '발생지시군구'], axis=1,inplace=True)
```

```python
df.to_csv('2012_2021_교통사망사고정보.csv',encoding='CP949')
```
